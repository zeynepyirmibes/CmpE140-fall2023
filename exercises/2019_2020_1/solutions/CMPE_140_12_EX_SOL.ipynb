# SAFETY LEVELS

## QUESTION

There is a built in data set named **USArrests** in RStudio. First column of this data set contains murder arrests per 100,000 residents in each of the 50 U.S. states in 1973.
Let's define 4 safety levels for US states as such :

|Murder Arrests(per 100,000)|Safety Level|
|---|---|
|2.5 and lower|Very Safe|
|2.5-7|Safe|
|7-15|Dangerous|
|15 and higher|Very Dangerous|

Now create function named **security** which takes 2 arguments:
* **dta** : a data frame which has **USArrests** as a default value
* **num** : any real number with no default value

Your function should ;
* Take the first column (The Murder column) of the data and create a factor from it using the classification above.
* Then, if **num** is negative, return a barplot. If not, return a pie chart as such :

~~~r
security(dta = USArrests, num = -2)
~~~



~~~r
security(dta = USArrests, num = 2)
~~~





## SOLUTION

````r
security <- function(dta,num)
  {
   #Creating factor using cut() function
    xy <- cut(dta[,1],
            breaks = c(-Inf, 2.5, 7,15, Inf),
            labels = c("Very Safe", "Safe", "Dangerous", "Very Dangerous"))
   t <- table(xy)
   
   if(num < 0)
   #Creating the barplot  
   {
        return(barplot(t,
                    main = "US Safety Levels"))
     }
   else {
   #Creating the pie chart
     return(pie(t,
                main = "US Safety Levels",
                col = rainbow(length(t))))
 }
   
 }

security(dta = USArrests, num = -2)

security(dta = USArrests, num = 2)
````

# BAT SIGNAL

## QUESTION

Gotham City is under the attack of 4 villains. You need to use the Bat Signal to summon Batman, but each villain stole a part of the signal and first you need to take those parts back and put them together to summon Batman and save the city!

Bat Signal parts that you took back from villains are expressed as these functions: 

```r
riddler <- function(x)
{
  abs(x/2)-0.0913722*(x^2)-3+sqrt(1-(abs(abs(x)-2)-1)^2)
}

penguin <- function(x)
{
    2*sqrt(-abs(abs(x)-1)*abs(3-abs(x))/((abs(x)-1)*(3-abs(x))))*(1+abs(abs(x)-3)/(abs(x)-3))*sqrt(1-(x/7)^2)+(5+0.97*(abs(x-0.5)+abs(x+0.5))-3*(abs(x-0.75)+abs(x+0.75)))*(1+abs(1-abs(x))/(1-abs(x)))
}

joker <- function(x)
{
  -3*sqrt(1-(x/7)^2)*sqrt(abs(abs(x)-4)/(abs(x)-4))
}

bane <- function(x)
{
  (2.71052+(1.5-0.5*abs(x))-1.35526*sqrt(4-(abs(x)-1)^2))*sqrt(abs(abs(x)-1)/(abs(x)-1))+0.9
}
```

1) Plot all four of these functions together to see which villain took which part of the signal as such:




2) Plot 4 functions on top of each other to form the Bat Signal as such: 



**Hints:** 
- You can use *par(mfcol = c(2, 2))* to set the number of graphs in the plot.
- You can use *par(new = True)* after a plot to add another plot on top of it.
- Setting the same xlim and ylim values for each plot will make it easier to fit them together.


## SOLUTION

````r

riddler <- function(x)
{
  abs(x/2)-0.0913722*(x^2)-3+sqrt(1-(abs(abs(x)-2)-1)^2)
}

penguin <- function(x)
{
    2*sqrt(-abs(abs(x)-1)*abs(3-abs(x))/((abs(x)-1)*(3-abs(x))))*(1+abs(abs(x)-3)/(abs(x)-3))*sqrt(1-(x/7)^2)+(5+0.97*(abs(x-0.5)+abs(x+0.5))-3*(abs(x-0.75)+abs(x+0.75)))*(1+abs(1-abs(x))/(1-abs(x)))
}

joker <- function(x)
{
  -3*sqrt(1-(x/7)^2)*sqrt(abs(abs(x)-4)/(abs(x)-4))
}

bane <- function(x)
{
  (2.71052+(1.5-0.5*abs(x))-1.35526*sqrt(4-(abs(x)-1)^2))*sqrt(abs(abs(x)-1)/(abs(x)-1))+0.9
}

#1
par(mfcol = c(2,2))
plot(bane, xlim = c(-7,7), ylim = c(-10,10))
plot(riddler, xlim = c(-7,7), ylim= c(-10,10))
plot(joker, xlim= c(-7,7), ylim = c(-10,10))
plot(penguin, xlim = c(-7,7), ylim = c(-10,10))

#2
par(mfcol = c(1,1))
plot(bane, xlim = c(-7,7), ylim = c(-10,10), xlab = "signal", ylab = "bat")
par(new = TRUE)
plot(riddler , xlim = c(-7,7), ylim = c(-10,10), xlab = "signal", ylab = "bat")
par(new = TRUE)
plot(joker, xlim= c(-7,7), ylim = c(-10,10), xlab = "signal", ylab = "bat")
par(new = TRUE)
plot(penguin, xlim = c(-7,7), ylim = c(-10,10), xlab = "signal", ylab = "bat")
````
