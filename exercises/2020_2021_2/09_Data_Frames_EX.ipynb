The data in the first row indicates that bank **a** offers 50.000 TL in monthly payments of 2467 TL with 1,4% interest rate per month for 24 months. 

A series of periodic payments that occur for some fixed numbers of periods are called annuities. Future value of an annuity can be calculated with the following formula:

$

**C :** constant payment

**r :** interest rate

**t :** number of periods
