Then you should:

2) When he buys something the amount should decrease from his bank account. He should not be able to buy an item that exceeds his budget. In that case, the function should return **"You do not have enough money"**.

*Hint: You can use* ``%in%`` *function to check whether an item is in the list*.

Here are some test scenarios for the **wallet** function:

## SOLUTION

````r
abdullatif <- list(bank_account =20000,
                 items_wanted = c("PS5","Laptop","Beach House","Fountain pen"),
                 items_owned = c("PS4","Iphone","TV","Headphones"))

abdullatif


wallet <- function(price,item,person=abdullatif)
{
    if(person$bank_account <price)
    {
        print("You do not have enough money")
    }
    else if(item %in% abdullatif$items_owned)
    {
        print("you already have this item")
    }
    else if (item %in% person$items_wanted ==F)
    {
        
        answer <-readline("This item is not in your wanted list. Do you really want it? (Yes or No):")
        if(answer == "Yes")
        {
            person$items_owned <- c(person$items_owned,item)
            person$bank_account <- person$bank_account - price
        }
    }
    else if (item %in% person$items_wanted ==T)
    {
        
        indice <-which(item %in% person$items_wanted)
        person$items_wanted <- person$items_wanted[-indice]
        person$items_owned <- c(person$items_owned, item)
        person$bank_account <- person$bank_account - price
        
    }
    
    return(person)
    
}
wallet(10000,"PS5")
wallet(500000,"Beach House")
wallet(5000,"Car")
````

## SOLUTION

````r
text <- "call me ishmael some years ago never mind how long precisely having little or no money in my purse and nothing particular to interest me on shore i thought i would sail about a little and see the watery part of the world it is a way i have of driving off the spleen and regulating the circulation whenever i find myself growing grim about the mouth whenever it is a damp drizzly november in my soul whenever i find myself involuntarily pausing before coffin warehouses and bringing up the rear of every funeral i meet and especially whenever my hypos get such an upper hand of me that it requires a strong moral principle to prevent me from deliberately stepping into the street and methodically knocking peoples hats off then i account it high time to get to sea as soon as i can this is my substitute for pistol and ball with a philosophical flourish cato throws himself upon his sword i quietly take to the ship there is nothing surprising in this if they but knew it almost all men in their degree some time or other cherish very nearly the same feelings towards the ocean with me"

#1
makelist <- function(txt)
{
  emp <- list()
  for (i in 1:length(strsplit(txt, split=" ")[[1]]))
    {
    emp[[i]] <- strsplit(txt, split=" ")[[1]][[i]]
  }
  return(emp)
}

#2
newlist <- lapply(text, makelist)
newlist <- newlist[[1]]
newlist

#3
textcheck <- function(wordset)
{
  wordnum <- list()
  for (i in 1:length(wordset))
    {
    if (is.null(wordnum[[wordset[[i]]]]))
      {
      wordnum[[wordset[[i]]]] <- 1
    } 
    else 
      {
        wordnum[[wordset[[i]]]] <- wordnum[[wordset[[i]]]] + 1 
    }
  }
  return(wordnum)
}

textcheck(newlist)
````

As you can see different type of variables are mixed. For example:

## SOLUTION

````r
#1
melon_data <- list(12,"Melon is something good", TRUE, 214, 54,"The Melon has a building.", "Consumer Graphs", 4i, FALSE,
          44, FALSE, 432, 11, TRUE,"I want to eat a Melon", "Consumer Behaviors", 2i, 
          "The Melon sold it.", 6i, "Consumer is not choosing.", FALSE, 654, "Consumer profile", 
          12i, FALSE,"The Melon is a well-known company", 222, "Consumer data", TRUE, 6i, 
          "The Melon company has been doing good", 21, 22i, FALSE, "I hate the Consumer")


#2

ordering <- function(lst, dep)
{
  int_data <- c()   # Create empty vectors so that you can save the data into them.
  int_mar <- c()
  cmp <- c()
  log <- c()
  char_data <- c()
  char_mar <- c() 
  
  for (i in 1:length(lst)) 
  {
  
    if(class(lst[[i]]) == "numeric")       
    {
      if(lst[[i]] > 100) {int_data <- c(int_data, lst[[i]])}
      
      else{int_mar <- c(int_mar, lst[[i]])}
    }
    
    else if(class(lst[[i]]) == "complex"){cmp <- c(cmp, lst[[i]])}
    
    else if(class(lst[[i]]) == "logical"){log <- c(log, lst[[i]])}
    
    else if(class(lst[[i]]) == "character")
    {
    
      sent <- lst[[i]]
      div <- strsplit(sent, " ")   # Use that function to split sentence into words. How many spaces you put
                                    # between "" makes the difference.
      
      if(div[[1]][2] == "Melon"){char_data <- c(char_data, lst[[i]])}
      
      else if(div[[1]][1] == "Consumer"){char_mar <- c(char_mar, lst[[i]])}
    }
    
      
  }
        

 # Create the lists for the two departments.
  
  lstm <- list(department="Marketing", numerics=int_mar, logicals=log, characters=char_mar)
  lstd <- list(department="Data", numerics=int_data, complex=cmp, characters=char_data)
  
  if(dep == "Marketing"){return(lstm)}   # Return the accurate list depending on the argument.
  else{return(lstd)}
}

ordering(melon_data, "Data")
    
ordering(melon_data, "Marketing")
````
