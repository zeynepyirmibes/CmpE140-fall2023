## Question 3 

GPA's of a graduating class of economics students follow a normal distribution, with a mean of 3.02 and a standart deviation of 0.28. There are 100 students in total. Draw a histogram of GPA's with 0.25 GPA per break. Make sure to show the limits of x axis clearly.

Set RNG version to 3.3.1 and seed to 1000.

*Hint 1* : You should use the *xlim* parameter to specify the x axis limits.

*Hint 2* : You can use use the *rnorm ()* function to create normally distributed random variables.

Your output should be:



## Solution 3

```r
RNGversion("3.3.1")
set.seed(1000)
gpa_list <- rnorm(100, mean = 3.02, sd = 0.28)
hist(gpa_list, breaks = (max(gpa_list) - min(gpa_list)) / 0.25, xlim = c(2,4))
```

## Question 4

Set RNG version to 3.3.1 and seed to 1907.

Generate a vector of 12 random integers between 0 and 100 (with replacement) that represent the number of baby rabbits born in each month over a one year period. Then make a barplot that displays the number of rabbit births in each month. 

*Hint*: You can use the parameter *name.args()* to display month labels.

Your output should look like this:



## Solution 4

```r
RNGversion("3.3.1")
set.seed(1907)
rabbit_births <- sample(1:100, 12, replace=TRUE)
months <- c("Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec")
barplot(rabbit_births, names.arg = months, xlab = "Month", ylab = "Number of Rabbit Births", main = "Rabbit Births by Month")
```

## Question 5

Imagine you are doing a research on the differences of life expectancy between males and females.

Life expectancy for both males and females is assumed to follow a uniform distribution.

Set RNG version to 3.3.1 and seed to 1980.

Generate two vectors of 30 random numbers between 0 and 100 and assign them to vectors **male** and **female**.

1. Show the relationship between male and female life expectancy on a scatter plot. Male and female life expectancies should be on the x-axis and y-axis respectively. Name the axes accordingly and title the plot as "Male and Female Life Expectancy". Represent males with blue circles and females with red squares. 

2. Visualise male and female life expectancy on a barplot side by side. Title of the plot should be the same.

*Hint 1*: You can use the function *runif ()* to create uniform random numbers.

*Hint 2*: You can use *main*, *xlab* and *ylab* arguments for labelling purposes.

*Hint 3*: You can use parameters *color* and *pch* to change the color and shape of observations.

*Hint 4*: You can use the function *cbind ()* to combine two vectors and the parameter *beside=T* to display two series side-by-side on a barplot.


Your outputs should look like these:



## Solution 5

```r
RNGversion("3.3.1")
set.seed(1980)
male <- runif(30,0,100)
female <- runif(30,0,100)

#1
plot(male, female, main = "Male and Female Life Expectancy", xlab = "Male", ylab = "Female", col = c("blue","red"), pch = c(16,15))

#2
data <- cbind(sort(male), sort(female))
barplot(data, beside=T, main="Male and Female Life Expectancy")
```
