## Question 4

Mehmet and Alper are working on an assignment for their data analysis project. They need to create a custom R function that helps them classify data into categories based on certain conditions. The data they have collected is related to student grades and attendance.

Write an R function called **classify_student** that takes two vectors as input: **grades** and **attendance**. The grades vector represents student grades as numeric values ranging from 0 to 100, and the attendance vector represents the number of days each student attended the class.

The function should classify each student's grade into one of the following categories:

* *F* for grades below 50.
* *D* for grades between 50 and 59 (50 included).
* *C* for grades between 60 and 69(60 included).
* *B* for grades between 70 and 79(70 included).
* *A* for grades 80 and above (80 included).

In addition to classifying grades, the function should also consider the attendance of each student.

If a student's grade is *F* but their attendance is 80% or higher, their grade should be upgraded to *D*.
If a student's grade is *A* but their attendance is less than 80%, their grade should be downgraded to *B*.
For all other cases, the grade remains the same.
The function should return a vector of the same length as the input with the corresponding letter grade for each student, taking attendance into account.

**Example outputs**

```r
1.
grades <- c(42, 85, 75, 91, 55)
attendance <- c(90, 72, 88, 76, 95)
classify_student(grades, attendance)
[1] "D" "B" "B" "B" "D"

2.
grades <- c(45, 70, 91, 56, 88)
attendance <- c(85, 70, 60, 75, 85)
classify_student(grades, attendance)
[1] "D" "B" "B" "D" "A"
```
