## Question 4

Professor Smith decided to create a function in R to analyze the IQ scores of students in his class. The functions purpose is to  classify the intelligence levels of the students based on their IQ scores.
Your task is to create a function called **iq_scale** in R that does the following:

1. Takes a vector of IQ scores as input.

2. Excludes the two lowest IQ scores and the two highest IQ scores from the vector.

3. Calculates the average IQ score from the remaining values.

Once you have the mean IQ score, the function should assign one of the following intelligence classifications:

* Below 114 (including 114): "Average intelligence"
* 115 to 129 (including 115 and 129) : "Above average or bright"
* 130 to 159 (including 130 and 159): "Highly gifted"
* Above 160 (including 160) : "Exceptionally gifted"

**Example outputs**

```r
sample_1 <- c(90,87,112,120,157,100,85,87,100,107)
iq_scale(sample_1)
[1] "Average intelligence"

sample_2 <- c(96,105,128,123,146,108,122,88,112,132)
iq_scale(sample_2)
[1] "Above average or bright"

sample_3 <- c(120,180,137,120,157,100,140,112,100,177)
iq_scale(sample_3)
[1] "Highly gifted"

sample_4 <- c(150,180,167,152,157,190,140,162,180,187)
iq_scale(sample_4)
[1] "Exceptionally gifted"
```
