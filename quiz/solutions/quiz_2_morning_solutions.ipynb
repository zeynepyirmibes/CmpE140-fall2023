## Question 1

Pablo the penguin is an artistic genius in Antarctica. He loves to create colorful paintings using his flippers. He has just finished his painting and wants to analyze the colors he used. We'll assign numbers from 1 to 7 to represent the frequency of each color in his paintings. 

Set RNG version to 3.3.1 and seed to 1960.

**a)** Sample 7 numbers between 1 and 10 with replacement, and set their names to the colors (respectively): Red, Blue, Green, Yellow, Orange, Purple, and Pink.

```
   Red   Blue  Green Yellow Orange Purple   Pink 
     5      9      4      6      4      4      8 
```

**b)** Print the names of the colors that he painted exactly 4 times. 


```
[1] "Green"  "Orange" "Purple"
```

````r
RNGversion("3.3.1")
set.seed(1910)
  
# a)
colors <- c("Red", "Blue", "Green", "Yellow", "Orange", "Purple", "Pink")
painting <- sample(1:10, 7, replace = TRUE)
names(painting)<-colors
painting

# b)
names(painting[painting == 4])
````

## Question 2

In a 2005 study, researchers tried to figure out the life expectancy of people. Total of 112 males and 98 females have been examined. Let's try to replicate the research on R. 

Steps:
1. Set RNGversion to "3.3.1" and seed to 25
2. Create a **histogram** for males, which consists of 112 uniformly distibuted numbers with a minimum value of 63 and a maximum value of 83. (The name of the histogram should be "Male Life Expectancy", the name of the x axis should be "Age", and the name of the y axis should be "Subjects")
3. Create a **histogram** for females, which consists of 98 uniformly distibuted numbers with a minimum value of 52 and a maximum value of 90. (The name of the histogram should be "Female Life Expectancy", the name of the x axis should be "Age", and the name of the y axis should be "Subjects")

* **Hint:** Use the function runif() to generate the uniformly distibuted numbers. 
* **Hint:** Use the function hist() to generate the histogram. 


Your output should look like this:<br>


````r
RNGversion("3.3.1")
set.seed(25)

males <- hist(runif(112, min=63, max=83), main="Male Life Expectancy", xlab="Age", ylab="Subjects")
females <- hist(runif(98, min=52, max=90), main="Female Life Expectancy", xlab="Age", ylab="Subjects")
````
